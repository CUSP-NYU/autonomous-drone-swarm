```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.animation as animation
from matplotlib import rc
rc('animation', html='jshtml')
import pandas as pd
```

### Note

For consistency with the report, we fliped the x and y coordinates and subtracted from 9 for y value.

### Position: Trained by Fixed Initial Position

Import trajectory data

```python
df = pd.read_csv('/content/drive/MyDrive/5_USI/10_Final_Report/Viz/trajectory/trajectory_pos_fixed.csv')
df.head()
```

```python
max_steps = len(df)
```

Visualization

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111)

ax.set_xlim((0, 10))
ax.set_ylim((0, 10))

# grid
ax.set_xticks(range(10))
ax.set_yticks(range(10))
plt.tick_params(labelbottom=False, labelleft=False)
plt.tick_params(color='white')
plt.grid(color='gray', linestyle='-', linewidth=1)

def init():
    # initialize an empty list of cirlces
    return []

def animate(i):
    frame_patches = []
    # mapping status
    for j in range(i):
        m1 = patches.Rectangle(xy=(df['drone_1_y'][j], 9-df['drone_1_x'][j]), width=1, height=1, fc='lightblue')
        m2 = patches.Rectangle(xy=(df['drone_2_y'][j], 9-df['drone_2_x'][j]), width=1, height=1, fc='darkseagreen')
        frame_patches.append(ax.add_patch(m1))
        frame_patches.append(ax.add_patch(m2))
    
    # drones
    d1 = patches.Circle(xy=(df['drone_1_y'][i] + 0.5, 9-df['drone_1_x'][i] + 0.5), radius=0.25, fc='steelblue')
    d2 = patches.Circle(xy=(df['drone_2_y'][i] + 0.5, 9-df['drone_2_x'][i] + 0.5), radius=0.25, fc='darkolivegreen')
    frame_patches.append(ax.add_patch(d1))
    frame_patches.append(ax.add_patch(d2))
    
    return frame_patches

anim = animation.FuncAnimation(fig, animate, init_func=init, frames=max_steps, interval=100, blit=True)
```

```python
anim
```
